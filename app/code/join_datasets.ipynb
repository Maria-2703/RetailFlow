# Joining datasets
---

## 1. Analyze raw data

```python
import pandas as pd

# read all of the csv previously generated and cleaned
cameras = pd.read_csv('./data/cameras.csv')
detections = pd.read_csv('./data/detections.csv')
sales = pd.read_csv('./data/sales.csv')
soldproducts = pd.read_csv('./data/soldproducts.csv')
stores = pd.read_csv('./data/stores.csv')
traffic = pd.read_csv("./data/traffic.csv")
zones = pd.read_csv('./data/zones.csv')
```

```python
# display all the tables
print("cameras.csv")
display(cameras.head())

print("detections.csv")
display(detections.head())

print("sales.csv")
display(sales.head())

print("soldproducts.csv")
display(soldproducts.head())

print("stores.csv")
display(stores.head())

print("traffic.csv")
display(traffic.head())

print("zones.csv")
display(zones.head())
```

## 2. Join dataset

### 2.1 Infrastructure

```python
# join cameras + zones `stores`
infrastructure = (
    cameras.merge(zones, on='zone_id', how='left')
           .merge(stores, on='store_id', how='left')  # left to include all cameras
           .rename(columns={
               'lims': 'camera_lims',
               'coord_lims': 'zone_lims',
               'condition': 'camera_condition'  # avoid confusions
           })
)

# reordering the columns
infrastructure = infrastructure[[
    'store_id', 'store_name', 'city',
    'zone_id', 'zone_name', 'zone_type', 'zone_lims',
    'camera_id', 'model','camera_condition','camera_lims', 
    'installation_date', 'm2', 'max_capacity'
]]

display(infrastructure.head())
```

```python
infrastructure.info()
```

This dataset provides a comprehensive technical audit of the retail network, transforming isolated hardware specs into a map that identifies exactly where the equipment is failing and confirms that the standardized store layouts are correctly supported by the right camera models across all Valencia locations.

### 2.2 Revenue

```python
#join soldproducts + sales + stores
revenue = (
    soldproducts.merge(sales, on='ticket_id', how='inner')
                .merge(stores, on='store_id', how='left')
                .rename(columns={
                    'name': 'product_name',
                    'category': 'product_category',
                    'price': 'product_price' # avoid confusions
                })
)

# reorder
revenue = revenue[[
    'ticket_id', 'timestamp',
    'store_id','store_name', 'city',
    'product_id', 'product_name', 'product_category', 'product_price',
    'total_euros', 'product_amount',
    'checkout_number', 'zone_id',
    'm2', 'max_capacity'
]]

display(revenue.head())
```

```python
revenue.info()
```

This is the backbone of retail analytics; it connects what was bought and where, allowing the stores to move beyond simple revenue totals to granular item-level performance. This dataset provides a better view of sales performance that identifies which product categories are top-sellers in specific cities and calculates the average transaction value relative to the physical capacity and checkout efficiency of each store location.

### 2.3 Validation

```python
# convert timestamps
detections['timestamp'] = pd.to_datetime(detections['timestamp'])
traffic['date_time'] = pd.to_datetime(traffic['date_time'])

# Fix: Lowercase 'h' for rounding
detections['hour_key'] = detections['timestamp'].dt.floor('h')
traffic['hour_key'] = traffic['date_time'].dt.floor('h')

# We join detections with cameras first to inherit the 'zone_id'
detections_with_zones = pd.merge(
    detections, 
    cameras[['camera_id', 'zone_id']], 
    on='camera_id', 
    how='left'
)

# now we can join with traffic because we have 'zone_id'
validation = pd.merge(
    detections_with_zones, 
    traffic, 
    on=['zone_id', 'hour_key'], 
    how='inner'
)

# Clean and reorder the validation dataset
validation = (
    validation.drop(columns=['store_id_y'])
              .rename(columns={'store_id_x': 'store_id'})
)

validation = validation[[
    'detection_id','tracking_id',
    'hour_key', 'timestamp', 'date_time', 'traffic_id',
    'store_id', 'zone_id', 'camera_id',
    'class_object',
    'visitor_count', 'peak_people',
    'confidence', 'coord_lims',
    'average_time_in_store', 'bounce_rate'
]]

display(validation.head())
```

```python
validation.info()
```

this join is critical for data validation because it cross-references the detections against traffic within the same hour and area.This is useful to validate sensor precision and identify discrepancies between computer vision and hardware-based visitor counts in specific zones.

## 3. Save new datasets

```python
infrastructure.to_csv('data/joined_datasets/infrastructure.csv', index=False)
revenue.to_csv('data/joined_datasets/revenue.csv', index=False)
validation.to_csv('data/joined_datasets/validation.csv', index=False)
```
