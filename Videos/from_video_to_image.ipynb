```python
import cv2
import os
```

```python
path = './Files/cam2/'
out_path = './Files/'
out_video_name = 'cam2.mp4'
out_video_full = out_path + out_video_name
```

```python
pre_imgs = os.listdir(path)
print(pre_imgs)
```

```python
img = []
for i in pre_imgs:

    i = path + i
    img.append(i)


print(img)
```

```python
frame = cv2.imread(img[0])
size = list(frame.shape)
del size[2]
size.reverse()
print(size)
```

```python
cv2_fourcc = cv2.VideoWriter_fourcc(*'mp4v')

video = cv2.VideoWriter(out_video_full, cv2_fourcc, 12, size)

for i in range(len(img)):

    video.write(cv2.imread(img[i]))
    print(f'Frame {i} of {len(img)}')

video.release()
print(f'Output video {out_path}')
```
